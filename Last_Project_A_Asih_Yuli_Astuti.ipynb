## Data Preparation

```
# Mengupload data ke Google Colab
import pandas as pd
from google.colab import files
file = files.upload()
```

```
#Menampilkan Data Frame
import io
df = pd.read_csv(io.StringIO(file['billboard_top_100.csv'].decode('utf-8')))
df
```

```
#Menampilkan data lagu yang berada di posisi pertama di setiap minggunya

df_rank_1 = df[df['rank'] == 1 ]
df_rank_1
```

```
# Mencari berapa kali lagu tersebut berada di posisi pertama
df_rank_1 = df_rank_1.groupby(["song"])
df_rank_1['rank'].sum()
```

## Data Visualization

```
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
song = ['All I Want For Christmas Is You', 'Blinding Lights', 'Cardigan', 'Circles', 'Dynamite', 'Franchise', 'Life Goes On', 'Mood', 'Positions', 'Rain On Me', 'Rockstar', 'Savage', 'Savage Love (Laxed - Siren Beat)', 'Say So', 'Stuck With U', 'The Box', 'The Scotts', 'Toosie Slide', 'Trollz', 'WAP', 'Watermelon Sugar']
jumlah = [3,4,1,1,3,1,1,6,1,1,7,1,1,1,1,11,1,1,1,4,1]
ax.bar(song,jumlah)
plt.show()
```

Dari data di atas, terlihat lagu dengan judul 'The Box' berada di posisi pertama dengan jumlah yang paling banyak . Itu artinya, lagu 'The Box' merupakan lagu yang paling dominan yang memuncaki Billboard pada tahun 2020 sehingga menjadikan lagu 'The Box' menjadi lagu paling popular selama tahun 2020.
