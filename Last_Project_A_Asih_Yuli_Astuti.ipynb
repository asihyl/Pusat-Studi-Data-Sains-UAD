```
#Upload data ke G.Colab
import pandas as pd
from google.colab import files
file = files.upload()
```

```
#Menampilkan Data Frame
import io
df = pd.read_csv(io.StringIO(file['billboard_top_100.csv'].decode('utf-8')))
df
```

```
#Menampilkan data lagu yang berada di posisi pertama di setiap minggunya

df_rank_1 = df[df['rank'] == 1 ]
df_rank_1
```

```
#Mencari berapa kali judul lagu tersebut berada di posisi pertama

df_a = df_rank_1[df_rank_1['song'] == 'All I Want For Christmas Is You']
df_b = df_rank_1[df_rank_1['song'] == 'Circles']
df_c = df_rank_1[df_rank_1['song'] == 'The Box']
df_d = df_rank_1[df_rank_1['song'] == 'Blinding Lights']
df_e = df_rank_1[df_rank_1['song'] == 'Toosie Slide']
df_f = df_rank_1[df_rank_1['song'] == 'The Scotts']
df_g = df_rank_1[df_rank_1['song'] == 'Say So']
df_h = df_rank_1[df_rank_1['song'] == 'Stuck With U']
df_i = df_rank_1[df_rank_1['song'] == 'Savage']
df_j = df_rank_1[df_rank_1['song'] == 'Rain On Me']
df_k = df_rank_1[df_rank_1['song'] == 'Rockstar']
df_l = df_rank_1[df_rank_1['song'] == 'Trollz']
df_m = df_rank_1[df_rank_1['song'] == 'Cardigan']
df_n = df_rank_1[df_rank_1['song'] == 'Watermelon Sugar']
df_o = df_rank_1[df_rank_1['song'] == 'WAP']
df_p = df_rank_1[df_rank_1['song'] == 'Dynamite']
df_q = df_rank_1[df_rank_1['song'] == 'Franchise']
df_r = df_rank_1[df_rank_1['song'] == 'Savage Love (Laxed - Siren Beat)']
df_s = df_rank_1[df_rank_1['song'] == 'Mood']
df_t = df_rank_1[df_rank_1['song'] == 'Positions']
df_u = df_rank_1[df_rank_1['song'] == 'Life Goes On']

df1 = df_a['rank'].count()
df2 = df_b['rank'].count()
df3 = df_c['rank'].count()
df4 = df_d['rank'].count()
df5 = df_e['rank'].count()
df6 = df_f['rank'].count()
df7 = df_g['rank'].count()
df8 = df_h['rank'].count()
df9 = df_i['rank'].count()
df10 = df_j['rank'].count()
df11 = df_k['rank'].count()
df12 = df_l['rank'].count()
df13 = df_m['rank'].count()
df14 = df_n['rank'].count()
df15 = df_o['rank'].count()
df16 = df_p['rank'].count()
df17 = df_q['rank'].count()
df18 = df_r['rank'].count()
df19 = df_s['rank'].count()
df20 = df_t['rank'].count()
df21 = df_u['rank'].count()
```

```
#Menampilkan data dalam bentuk dataframe

data = {
    'song' : ['All I Want For Christmas Is You', 'Circles', 'The Box', 'Blinding Lights', 'Toosie Slide', 'The Scotts', 'Say So', 'Stuck With U', 'Savage', 'Rain On Me', 'Rockstar', 'Trollz', 'Cardigan', 'Watermelon Sugar', 'WAP', 'Dynamite', 'Franchise', 'Savage Love (Laxed - Siren Beat)', 'Mood', 'Positions', 'Life Goes On'],
    'jumlah' : [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16, df17, df18, df19, df20, df21]
}

data = pd.DataFrame(data,columns=['song','jumlah'])
print(data)
```

## Data Visualization

```
import matplotlib.pyplot as plt

%matplotlib inline

fig = plt.figure(figsize=(7,3))
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('song')
ax.set_ylabel('jumlah')

ax.bar(data['song'],data['jumlah'])
plt.show()
```

Dari data di atas, terlihat lagu dengan judul 'The Box' berada di posisi pertama dengan jumlah yang paling banyak . Itu artinya, lagu 'The Box' merupakan lagu yang paling dominan yang memuncaki Billboard pada tahun 2020.
